```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:

import requests
import pandas as pd
# In[ ]:

api_base_url = "https://jsonplaceholder.typicode.com" # Actually it is the documentation page
# Users endpoint to consume
users_endpoint = "{}/users".format(api_base_url)
# In[ ]:

# Let's perform an HTTP GET request, the response will be a JSON array
response = requests.get(users_endpoint)
print(response.headers.get('Content-Type'))
users = response.json()
user_1 = users[0]

# In[ ]:

response = requests.get("https://jsonplaceholder.typicode.com/posts")
posts = response.json()

response = requests.get("https://jsonplaceholder.typicode.com/comments")
comments = response.json()
```

```python
id_to_post = {post["id"]:post for post in posts}
for comment in comments:
    id_post = comment["postId"]
    if "comment" not in id_to_post[id_post]:
        id_to_post[id_post]["comment"] = []
    id_to_post[id_post]["comment"].append(comment["body"])
```

```python
# iterate over list of users and join it with other data like posts
user_to_post = {}
user_to_name = {}
for user in users:
    user_to_post[user["id"]] = [] 
    user_to_name[user["id"]] = user["name"] 
for id_post in id_to_post:
    user_to_post[id_to_post[id_post]["userId"]].append({
        "name": user_to_name[user["id"]],
        "title": id_to_post[id_post]["title"],
        "body": id_to_post[id_post]["body"],
        "comment": id_to_post[id_post]["comment"]
    })
```

```python
final_posts = [ 
    post
    for id_user in user_to_post
    for post in user_to_post[id_user]
]
```
